```python
import os
import sys
import json
root_path = os.path.abspath(os.path.join(os.getcwd(), "../.."))
sys.path.append(root_path)
import logging
from typing import List, Tuple, Dict

ROOT_DIR = r"C:\Projects\Research\SWEEP\SWEEP"

# Set up logging
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

def get_sectors(train=True) -> List[str]:
    data_path = os.path.join(ROOT_DIR, "data", f'{"train" if train else "test"}')
    return [sector for sector in os.listdir(data_path) if os.path.isdir(os.path.join(data_path, sector))]

def get_activities(sector: str, train=True) -> List[str]:
    sector_path = os.path.join(ROOT_DIR, "data", f'{"train" if train else "test"}', sector)
    return [activity for activity in os.listdir(sector_path) if os.path.isdir(os.path.join(sector_path, activity))]

from src.bpmn_generation import BPMNGenerationEvaluation
bpmn_evaluator = BPMNGenerationEvaluation()
```

```python
TRAIN = False

failures = {}
for sector in get_sectors(train=TRAIN):
    for activity in get_activities(sector, train=TRAIN):
        
        # if activity != "purchase_order":
        #     continue
        try:
            # save_path = os.path.join(ROOT_DIR, f"test\\results\\bpmn_generation\\test\\{sector}\\{activity}")
            
            # # create folder if it does not exist
            # if not os.path.exists(save_path):
            #     os.makedirs(save_path)
                
            # with open(os.path.join(save_path, f"{activity}_bpmn_output.json"), 'r') as f:
            #     data = json.load(f)
            #     if data["equivalent"]:
            #         continue
                
            #     continue
                
            print(f"Generating BPMN for [{sector}]-[{activity}]")
            
            final_equivalence, final_mermaid_code, final_revisions, num_iterations = bpmn_evaluator.generate_BPMN(sector, activity, train=TRAIN)
            save_output = {
                "equivalent": final_equivalence,
                "num_iterations": num_iterations,
                "final_mermaid_code": final_mermaid_code
            }
            
            save_path = os.path.join(ROOT_DIR, f"test\\results\\bpmn_generation\\{'train' if TRAIN else 'test'}\\{sector}\\{activity}")
            # create folder if it does not exist
            if not os.path.exists(save_path):
                os.makedirs(save_path)
                
            with open(os.path.join(save_path, f"{activity}_bpmn_output.json"), 'w') as f:
                json.dump(save_output, f)
        except Exception as e:
            failures[f"[{sector}]-[{activity}]"] = str(e)
```
