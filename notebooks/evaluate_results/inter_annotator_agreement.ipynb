## Interannotator Agreement


```python
from src.value_adding_analysis.inter_annotator_aggrement import InterAnnotatorAgreement

agreement_calculator = InterAnnotatorAgreement()

# Add annotations to the calculator
agreement_calculator.add_annotations('b2b_sales_will', 'Will', b2b_will_analysis)
agreement_calculator.add_annotations('b2b_sales_will', 'Abel', b2b_abel_analysis)
# agreement_calculator.add_annotations('b2b_sales_student', 'Student', b2b_student_analysis)
agreement_calculator.add_annotations('b2b_sales_student', 'Will', b2b_will_relabelled_analysis)
agreement_calculator.add_annotations('b2b_sales_student2', 'Student 2', b2b_student2_analysis)
agreement_calculator.add_annotations('b2b_sales_student2', 'Will', b2b_will_relabelled2_analysis)
# agreement_calculator.add_annotations('special_order_will', 'Student', special_order_student_analysis)
agreement_calculator.add_annotations('special_order_will', 'Student 2', special_order_student2_analysis)
agreement_calculator.add_annotations('special_order_will', 'Will', special_order_will_analysis)

# Compute Krippendorff's Alpha
alpha = agreement_calculator.compute_krippendorff_alpha()
print(f"Krippendorff's Alpha: {alpha:.4f}",end="\n=================================\n")

# Compute pairwise Cohen's Kappa
kappa_scores = agreement_calculator.compute_pairwise_cohen_kappa()
for annotator_pair, kappa in kappa_scores.items():
    print(f"Cohen's Kappa between {annotator_pair[0]} and {annotator_pair[1]}: {kappa:.4f}")

print("=================================")
# Compute pairwise MCC
mcc_scores = agreement_calculator.compute_pairwise_mcc()
for annotator_pair, mcc in mcc_scores.items():
    print(f"MCC between {annotator_pair[0]} and {annotator_pair[1]}: {mcc:.4f}")

print("=================================")

# Compute pairwise percent agreement
agreement_scores = agreement_calculator.compute_pairwise_percent_agreement()
for annotator_pair, agreement in agreement_scores.items():
    print(f"Percent agreement between {annotator_pair[0]} and {annotator_pair[1]}: {agreement*100:.2f}%")
```

```python
# Compute precision, recall, and F1-score for NVA
metrics = agreement_calculator.compute_precision_recall_f1(target_class='NVA')
for annotator, scores in metrics.items():
    print(f"Annotator: {annotator}")
    print(f"  Precision (NVA): {scores['precision']:.4f}")
    print(f"  Recall (NVA): {scores['recall']:.4f}")
    print(f"  F1-Score (NVA): {scores['f1_score']:.4f}",end="\n=================================\n")
```

```python
# Compute confusion matrices
confusion_matrices, categories = agreement_calculator.compute_pairwise_confusion_matrices()
for annotator_pair, cm in confusion_matrices.items():
    print(f"Confusion Matrix between {annotator_pair[0]} and {annotator_pair[1]}:")
    print("Rows: Actual labels from first annotator")
    print("Columns: Predicted labels from second annotator")
    print(f"Categories: {categories}")
    print(cm,end="\n=================================\n")
```

# Top Model Interannotator Agreement


```python
from src.value_adding_analysis.inter_annotator_aggrement import InterAnnotatorAgreement

import logging
from typing import List, Tuple, Dict

ROOT_DIR = r"C:\Projects\Research\SWEEP\SWEEP"

# Set up logging
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

def get_sectors(train=True) -> List[str]:
    data_path = os.path.join(ROOT_DIR, "data", "train" if train else "test")
    return [sector for sector in os.listdir(data_path) if os.path.isdir(os.path.join(data_path, sector))]

def get_activities(sector: str, train=True) -> List[str]:
    sector_path = os.path.join(ROOT_DIR, "data", "train" if train else "test", sector)
    return [activity for activity in os.listdir(sector_path) if os.path.isdir(os.path.join(sector_path, activity))]

def get_file_paths(sector: str, activity: str, model_name: str, train=True) -> Tuple[str, str, str, str]:
    data_path = os.path.join(ROOT_DIR, "data", "train" if train else "test", sector, activity)
    test_path = os.path.join(ROOT_DIR, "test", "results", "value_adding_analysis", sector, activity)
    ground_truth_path = os.path.join(ROOT_DIR, data_path, f"{activity}_step_value_analysis.json")
    model_response_path = os.path.join(ROOT_DIR, test_path, f"{model_name}_response.json")
    return test_path, ground_truth_path, model_response_path

agreement_calculator = InterAnnotatorAgreement()


def add_annotations(sector: str, activity: str, model_name: str, train=True) -> None:
    test_path, ground_truth_path, model_response_path = get_file_paths(sector, activity, model_name, train)
    ground_truth = open_file(ground_truth_path, ".json")
    model_response = open_file(model_response_path, ".json")
    agreement_calculator.add_annotations(f"{sector}_{activity}", "Ground Truth", ground_truth)
    agreement_calculator.add_annotations(f"{sector}_{activity}", model_name, model_response['response'])


```

```python
model = "GPT-3.5-Value-Classification-Lean_Expert_Detailed"
def main():
    count = 0
    sectors = get_sectors()
    print(sectors)
    for sector in sectors:
        activities = get_activities(sector)
        for activity in activities:
            try:
                add_annotations(sector, activity, model, train=True)
                count += 1
            except:
                pass
    return count


if __name__ == "__main__":
    count = main()
    print(f"Added {count} annotations")
```
