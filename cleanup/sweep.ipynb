```python
from main import main
from pprint import pprint
import os

SECTOR = "customer_service"
ACTIVITY = "rental_request"
DIR_PATH = os.path.join("data", SECTOR, ACTIVITY)

PROCESS = f"{ACTIVITY}_model.txt"

def read_process(dir_path, process):
    with open(os.path.join(dir_path, process), "r") as file:
        text = file.read()
    return text

GPT_NAME = "GPT-3.5-Value-Model"
```

```python
process = read_process(DIR_PATH, PROCESS)
    
activity_breakdown, value_classification = main(process, step_level=True, verbose=True)
```
